```python
import requests
from lxml import etree
import pandas as pd
pd.set_option('display.max_rows', 500)
```

```python
url = "https://www.technofino.in/community/threads/best-credit-cards-2024-category-wise-credit-card-recommendation-by-technofino.21160/"
```

```python
response = requests.get(url)
```

```python
header_xpath = '//div[@data-lb-id="post-470665"]//span[(@style="font-size: 18px")]//text()'
```

```python
dom = etree.HTML(response.content)
headers = dom.xpath(header_xpath)
```

```python
fixed_headers = []
for i in range(len(headers)-1, -1, -1):
  if not headers[i].startswith("Best"):
    headers[i-1] += headers[i]
  else:
    fixed_headers = [headers[i].replace("Best Credit Cards For ", "").replace(":","")] + fixed_headers
```

```python
fixed_headers
```

```python
dfs = pd.read_html(url)
```

```python
for df, cat in zip(dfs, fixed_headers):
  df["Category"] = cat
  if "Effective Reward" in df.columns:
    df['Reward'] = df[[col for col in df.columns if "Effective Reward" in col][0]]
```

```python
cc = pd.concat(dfs)[dfs[0].columns]
cc = cc.dropna(subset=["Card Name"])
cc['Card Type'].fillna(method='ffill', inplace=True)
```

```python
cc_groups = cc.groupby(['Card Type', "Card Name", 'Card\'s Reward Type', 'Category', 'Reward', 'Capping']).agg("count")
```
